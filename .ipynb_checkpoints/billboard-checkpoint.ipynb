```python
import json
import billboard
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# read creditals (Client ID and Client Secret) from file
creditals = []
f = open('creditals.txt', 'r')
for line in f:
	#add line to creditals list (remove trailing whitespace characters)
	creditals.append(line.rstrip())

# create Spotify object with creditals
client_credentials_manager = SpotifyClientCredentials(creditals[0], creditals[1])
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

print sp
```

```python
#club 27 members with suicide date: https://en.wikipedia.org/wiki/27_Club
club27members = dict()

#club27members[''] = ''
club27members['Kurt Cobain'] = '1994-04-05'
club27members['Amy Winehouse'] = '2011-07-23'
club27members['Jimi Hendrix'] = '1970-09-18'
#club27members['Janis Joplin'] = '1970-10-04'
#club27members['Jim Morrison'] = '1971-07-03'


print club27members
```

```python
import operator

songsToSearch = 10

memberGenres = dict()
memberArtists = dict()
artistGenres = dict()
allGenres = []
allArtists = []

def addGenreToPopularityDict(genresDict, genre):
    if (genresDict.has_key(genre)):
        #increase popularity
        genresDict[genre] += 1
    else:
        #add popularity value
        genresDict[genre] = 1
        
def addGenresToArtistDict(artist, genres):
    if not artist in artistGenres.keys():
        artistGenres[artist] = genres
    
        
for k,v in club27members.items():
    genresPopularity = dict()
    billboardArtists = []
    sArtists = 0
    fArtists = 0
    processedSongs = 0
    print 'Getting Billboard Data for ' + k
    chart = billboard.ChartData('hot-100', date=v)

    for song in chart:
        if (processedSongs<songsToSearch):
            #print(('{} : ' + song.title + ' by ' + song.artist).format(song.rank))
            #split multiple artist
            for singleArtist in song.artist.replace('Featuring', '&').replace(',', '&').split('&'):
                stripedSingleArtist = singleArtist.lstrip().rstrip()
                billboardArtists.append(stripedSingleArtist)
                allArtists.append(stripedSingleArtist)
                #print 'Searching for: ' + stripedSingleArtist
                foundArtists = sp.search(q='artist:%'+stripedSingleArtist, type='artist')['artists']['items']
                #print (json.dumps(foundArtists, indent=2))
                for foundArtist in foundArtists:
                    if (foundArtist['name'] == stripedSingleArtist and foundArtist['popularity'] > 50):
                        #print 'Found match for: ' + stripedSingleArtist
                        fArtists += 1
                        addGenresToArtistDict(stripedSingleArtist, foundArtist['genres'])
                        #genres 
                        for genre in foundArtist['genres']:
                            addGenreToPopularityDict(genresPopularity, genre)
                            allGenres.append(genre)
                sArtists += 1

            processedSongs += 1
    print('artists: {} found / {} searched'.format(fArtists, sArtists))
    #sortedGenresPopularity = sorted(genresPopularity.items(), key=operator.itemgetter(1), reverse=True)
    #print sortedGenresPopularity
    memberGenres[k] = genresPopularity
    memberArtists[k] = billboardArtists
print memberGenres
print artistGenres
```

```python
import numpy as np
from matplotlib import pyplot as plt

#get all unique genres
allUniqueGenres = np.unique(allGenres)
#print allUniqueGenres
#print len(allUniqueGenres)

ind = np.arange(len(allUniqueGenres))# the x locations for the groups
width = 0.35       # the width of the bars

#fig = plt.figure()
fig, ax = plt.subplots()
plt.figure(figsize=(1,1))

allRects = []

#for all 27 member
noM = 0
for k,v in memberGenres.items():
    values = []
    #get data value for all found genres
    for g in allUniqueGenres:
        if v.has_key(g):
            values.append(v[g])
        else:
            values.append(0)
    #print values
    allRects.append(ax.bar(ind + noM * width, values, width))
    noM +=1
    
    

#plt.bar(range(len(allUniqueGenres)), values)
ax.set_xticks(ind+width/(noM+1))
ax.set_xticklabels(allUniqueGenres, rotation=90)
ax.legend(allRects, memberGenres.keys())

plt.show()
```
